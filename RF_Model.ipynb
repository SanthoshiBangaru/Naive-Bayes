```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)

categorical_cols = ['city_tier', 'department', 'education_level']

ohe_array = ohe.fit_transform(df[categorical_cols])

ohe_df = pd.DataFrame(
    ohe_array,
    columns=ohe.get_feature_names_out(categorical_cols),
    index=df.index
)

df = pd.concat([df.drop(columns=categorical_cols), ohe_df], axis=1)

df.head()
```

```python
ohe_df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['target_left_company'])
y = df['target_left_company']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)

print("Accuracy : ",accuracy_score(y_test,rf_pred))
print("Confusion Matrix : \n",confusion_matrix(y_test,rf_pred))
print("classfication Report :\n ",classification_report(y_test,rf_pred))
```
