### Naive Bayes ML Algorithm Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
x, y = load_iris(return_X_y=True)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train
```

```python
from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB()
bnb = model.fit(x_train, y_train)
bnb_pred = bnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Bernoulli:")
print("Accuracy Score:", accuracy_score(y_test, bnb_pred))
print("Classification Report:", classification_report(y_test, bnb_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, bnb_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
gnb = model.fit(x_train, y_train)
gnb_pred = gnb.predict(x_test)

print("Gaussian:")
print("Accuracy Score:", accuracy_score(y_test, gnb_pred))
print("Classification Report:", classification_report(y_test, gnb_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, gnb_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
mnb = model.fit(x_train, y_train)
mnb_pred = mnb.predict(x_test)

print("Multinomial:")
print("Accuracy Score:", accuracy_score(y_test, mnb_pred))
print("Classification Report:", classification_report(y_test, mnb_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, mnb_pred))
```
